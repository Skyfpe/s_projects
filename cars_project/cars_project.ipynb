# Функции для проекта

```python
def unique_identifier(df):
    print('#########################')
    for column_name in df.select_dtypes(include='object').columns:
            print(f'unique values in col "{column_name}": {df[column_name].unique()}')
            print('#########################')
```

```python
def bootstrap_mean_dist(df, col, num_subsamples, n):
    
    state = np.random.RandomState(RANDOM_STATE)
    values = []
    
    for i in range (num_subsamples):
        
        subsamp = df[col].sample(n=n, replace=False, random_state=state)
        values.append(subsamp.mean())
        
    return values
```

```python
def num_cols_bootstrap_stattest():
    
    cars_na = cars[cars.isna().any(axis=1)]
    
    for column in cars.select_dtypes(include=['int64', 'float64']).columns:
        
        not_na = np.array(bootstrap_mean_dist(cars.dropna(axis=0), column, 1000, 10000))
        na = np.array(bootstrap_mean_dist(cars_na, column, 1000, 10000))
        
        result = stats.ttest_ind(not_na, na, alternative='two-sided')
        
        if result.pvalue < alpha:
            print('Похоже, статистически значимая разница между распределениями средних признака "{}" есть'.format(column))
        else:
            print('Похоже, статистически значимой разницы между распределениями средних признака "{}" нет'.format(column))
```

```python
def num_cols_stattest_ind(samp_1, samp_2):
    
    for column in samp_1.select_dtypes(include=['int64', 'float64']).columns:
        
        result = stats.ttest_ind(samp_1[column], samp_2[column], alternative='two-sided')
        
        if result.pvalue < alpha:
            print('Похоже, статистически значимая разница между распределениями признака "{}" есть'.format(column))
        else:
            print('Похоже, статистически значимой разницы между распределениями средних признака "{}" нет'.format(column))
```

```python
def residuals_analysis_plotter(y_true, y_pred, figsize=(15, 5), xlabel='target_feature', mean_text_x_coef_degree=5, mean_text_y_coef=2, bins=50, suptitle=None, scatter_grid=False):
    
    residuals = y_true - y_pred
    
    fig, ax = plt.subplots(1, 2, figsize=figsize)
    
    ax[0].hist(residuals, bins=bins, color='silver', alpha=0.8)
    ax[0].set_title('распределение остатков (ошибок) модели')
    ax[0].set_xlabel('errors', size=13)
    ax[0].set_ylabel('count', size=13)
    ax[0].grid()
    
    ax[0].axvline(residuals.mean(), color='C3', linestyle='dashed', linewidth=2.15)
    ax[0].text(x=residuals.mean() + 2**mean_text_x_coef_degree, y=ax[0].get_yticks().max() / mean_text_y_coef, s='Остаточное среднее: {:.2f}'.format(residuals.mean()), size=10)
    
    ax[1].scatter(y_pred, residuals)
    
    ax[1].set_title('дисперсия остатков (ошибок) модели')
    ax[1].set_xlabel('{}'.format(str(xlabel)), size=13)
    ax[1].set_ylabel('цена ошибки (error)', size=13)
    
    if scatter_grid == True:
        ax[1].grid()
    else:
        pass
    
    if suptitle != None:
        fig.suptitle(suptitle, fontsize=15)
    else:
        pass
        
    plt.show()
```

```python
def rel_prioritized_inversed_neg_scoring(df_with_ordered_cols, competitors, exclude_cols=None):
    
    df = df_with_ordered_cols.copy()
    
    if exclude_cols == None:
        sup_df = df
    else:
        sup_df = df.drop(exclude_cols, axis=1)
        
    degree = len(sup_df.select_dtypes(include=['int64', 'float64']).columns)
    
    for column in sup_df.select_dtypes(include=['int64', 'float64']).columns:
        
        score_name = '{}_score^{}'.format(column, degree)

        new_df = sup_df[[competitors, column]].rename(columns={column: score_name})

        new_df[score_name] = new_df[score_name] + 1

        new_df[score_name] = round(((-new_df[score_name] + new_df[score_name].max()) / (new_df[score_name].max() - new_df[score_name].min())) * np.exp(degree), 2)

        degree -= 1

        df = df.merge(new_df, on=competitors)
    
    total_score = df.iloc[:, -len(sup_df.select_dtypes(include=['int64', 'float64']).columns):].sum(axis=1)
    
    df['score_total'] = total_score
    
    df = df.sort_values(by='score_total', ascending=False).reset_index(drop=True)
    
    return df
```

Инициализируем параметры для моделей. Будем использовать три: попробуем дерево решений, случайный лес и линейную регрессию с L2-регуляризацией ввиду большой размерности признакового пространства после кодирования категориальных признаков.

### Проверка на адекватность и анализ качества работы выбранной модели

Сначала проверим выбранную модель на адекватность. Для этого подберем параметры и найдем лучший DummyRegressor, который всегда предсказывает константу.

```python
dummy_pipe = Pipeline(
    [
        ('preprocessing', data_preprocessing),
        ('models', DummyRegressor())
    ]
)

dummy_params = [
    {
        'models': [DummyRegressor()],
        'models__strategy': ['mean',
                             'median'
                            ],
        'preprocessing__num': [StandardScaler(),
                               MinMaxScaler(),
                               RobustScaler(),
                               'passthrough'
                              ]
    },
    {
        'models': [DummyRegressor()],
        'models__strategy': ['quantile'],
        'models__quantile': np.arange(0, 1, 0.05),
        'preprocessing__num': [StandardScaler(),
                               MinMaxScaler(),
                               RobustScaler(),
                               'passthrough'
                              ]
    },
]

dummy_grid = GridSearchCV(dummy_pipe,
                          param_grid=dummy_params,
                          cv=3,
                          scoring='neg_root_mean_squared_error',
                          n_jobs=-1,
                          verbose=2
                         )

dummy_grid.fit(X_train, y_train)
```

```python
print(f'Лучшее значение RMSE при кросс_валидации: {round(np.abs(dummy_grid.best_score_), 2)}')
print(f'Лучшая dummy-модель и её параметры: {dummy_grid.best_params_}')
```
